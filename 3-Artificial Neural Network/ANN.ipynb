### Size of layers and initializing parameters weights and bias

Weight'in başlangıç değerine neden 0 vermiyoruz?: Çünkü 0 değeri ile başlatırsak belli bir iterasyondan sonra gradient descent artık farklılık yaratmamaya başlayacak, hep aynı şeyi hesaplamaya başlayacak ve bu durumda linear regression'dan hiçbir fark olmayacaktır ve çeşitlilik elde edilemeyecektir.

Değerleri küçük tanımlama sebbeimiz ise, tanh fonksiyonuna verilen değerler eğer büyük olursa güncelleme süresi çok uzar ve bu yüzden weight'e büyük sayılar veremiyoruz. 0.01 civarında küçük sayıları weight'e atayacağız.

```python
# intialize parameters and layer sizes
def initialize_parameters_and_layer_sizes_NN(x_train, y_train):
    parameters = {"weight1": np.random.randn(3,x_train.shape[0]) * 0.1,
                  "bias1": np.zeros((3 ,1)),
                  "weight2": np.random.randn(y_train.shape[0],3) * 0.1,
                  "bias2": np.zeros((y_train.shape[0],1))}
    return parameters
```

Bu 3 sayısı 3 tane node olduğunu temsil ediyor. Bu yüzden np.random.randn(3,x_train.shape[0]) * 0.1 satırında weight'in row sayısı 3 olarak belirlenmiştir çünkü 3 tane node vardır. Tüm bu anlattıklarım z=w*x+b formülüne göredir.

z=w*x+b formülüne göre mantıken w*x kısmı 3,1'dir ve bununla bir şeyi toplayabilmemiz için size'larının aynı olması gerekmektedir dolayısıyla bias'ta 3,1 olmalıdır.

### Forward Propagation

ileriye doğru pikseller ve weight'ler çarpılır ve bias ile toplanır. z değeri elde edilir ve z değeri tanh fonksiyonuna verilir. Daha sonra hidden layer'ın içeriği elde edilir. Daha sonra hidden layer'da weight ile çarpılır ve daha sonra bias ile toplanır. z2 değeri elde edilir ve bu değer sigmoid fonksiyonuna verilir.

```python
def forward_propagation_NN(x_train, parameters):

    Z1 = np.dot(parameters["weight1"],x_train) +parameters["bias1"]
    A1 = np.tanh(Z1)
    Z2 = np.dot(parameters["weight2"],A1) + parameters["bias2"]
    A2 = sigmoid(Z2)

    cache = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}
    
    return A2, cache
```

## Loss and Cost Functions

```python
# Compute cost
def compute_cost_NN(A2, Y, parameters):
    logprobs = np.multiply(np.log(A2),Y)
    cost = -np.sum(logprobs)/Y.shape[1]
    return cost
```

logprobs = np.multiply(np.log(A2),Y) satırında, A2 ile yani output'taki y_head değeri ile Y'yi yani gerçek değerleri logaritmik bir şekilde çarpıyoruz.

## Backward Propagation

Şimdi weight ve bias değerlerini güncellemeliyiz. Bunun için backward propagation yapmalıyız. Yani cost fonksiyonunun weight'lere göre türevini bulup weight'leri bu türevler doğrultusunda belli bir kural içerisinde güncelleyeceğiz.

```python
# Backward Propagation
def backward_propagation_NN(parameters, cache, X, Y):

    dZ2 = cache["A2"]-Y  #cost'un z2'ye göre türevidir.
    dW2 = np.dot(dZ2,cache["A1"].T)/X.shape[1]
    db2 = np.sum(dZ2,axis =1,keepdims=True)/X.shape[1]
    dZ1 = np.dot(parameters["weight2"].T,dZ2)*(1 - np.power(cache["A1"], 2))  #cost'un Z1'e göre türevi demektir.
    dW1 = np.dot(dZ1,X.T)/X.shape[1]  #cost'un w1'e göre türevi demektir.
    db1 = np.sum(dZ1,axis =1,keepdims=True)/X.shape[1]
    grads = {"dweight1": dW1,
             "dbias1": db1,
             "dweight2": dW2,
             "dbias2": db2}
    return grads
```

def backward_propagation_NN(parameters, cache, X, Y): satırındaki parametreleri açıklayalım. parameters isimli parametre ile weight1, weight2, bias1, bias2 değişkenlerini alacağız. X resimler ve Y label değerleridir.

Cost fonksiyonu y_head kısmındadır yani outputtadır. Cost fonksiyonundan weight1 'e gidebilmek için cost'un z2'ye göre türevi, z2'ninde w1'e göre türevini almalıyız. Bias içinde aynı şey geçerlidir. Yani cost'un z2'ye göre türevini ve z2'ninde z1'e göre türevini bulursak cost'un w1'e göre olan değişimini bulmuş olabiliriz.

keepdims=True parametresi ile array olarak tutulması sağlanmıştır. Yani eğer sum işleminin sonucu sabit bir sayı olsa bile bunu array'in içerisine yazıyor.

grads içerisine dweight1, dweight2, dbias1, dbias2 'deki değişimler depolanıyor. Yani cost'un bunlara göre olan türevleri depolanıyor. Bunları return eddeceğiz ve update metodunda kullanacağız.

## Update Parameters

```python
# update parameters
def update_parameters_NN(parameters, grads, learning_rate = 0.01):
    parameters = {"weight1": parameters["weight1"]-learning_rate*grads["dweight1"],
                  "bias1": parameters["bias1"]-learning_rate*grads["dbias1"],
                  "weight2": parameters["weight2"]-learning_rate*grads["dweight2"],
                  "bias2": parameters["bias2"]-learning_rate*grads["dbias2"]}
    
    return parameters
```

Parametreler update edilip öğrenildiğine göre model hazırdır dolayısıyla artık prediction yapabiliriz.

## Prediction

```python
# prediction
def predict_NN(parameters,x_test):
    # x_test is a input for forward propagation
    A2, cache = forward_propagation_NN(x_test,parameters)
    Y_prediction = np.zeros((1,x_test.shape[1]))
    # if z is bigger than 0.5, our prediction is sign one (y_head=1),
    # if z is smaller than 0.5, our prediction is sign zero (y_head=0),
    for i in range(A2.shape[1]):
        if A2[0,i]<= 0.5:
            Y_prediction[0,i] = 0
        else:
            Y_prediction[0,i] = 1

    return Y_prediction
```

A2, cache = forward_propagation_NN(x_test,parameters) satırında öncelikle forward propagation ile y_head değeri elde edilir. A2, y_head değeridir. 

Y_prediction = np.zeros((1,x_test.shape[1])) satırında tahminler için yer ayırıyoruz.

Alttaki for içerisinde yapılan ise, eğer A2 değeri 0.5'ten küçük ise resimdeki işaret 0'dır,  A2 değeri 0.5'ten büuük ise resimdeki işaret 1'dir. Bu bilgileri Y_prediction array'inin içerisinde depoluyoruz. Sonrasında bu array return ediliyor.

Şimdi bu zamana kadar yazdığımız tüm metodları birleştirerek modeli oluşturalım.

## Create ANN Model

```python
# calculation of z
#z = np.dot(w.T,x_train)+b
def sigmoid(z):
    y_head = 1/(1+np.exp(-z))
    return y_head
```

```python
# 2 - Layer neural network
def two_layer_neural_network(x_train, y_train,x_test,y_test, num_iterations):
    cost_list = []
    index_list = []
    #initialize parameters and layer sizes
    parameters = initialize_parameters_and_layer_sizes_NN(x_train, y_train)

    for i in range(0, num_iterations):
         # forward propagation
        A2, cache = forward_propagation_NN(x_train,parameters)
        # compute cost
        cost = compute_cost_NN(A2, y_train, parameters)
         # backward propagation
        grads = backward_propagation_NN(parameters, cache, x_train, y_train)
         # update parameters
        parameters = update_parameters_NN(parameters, grads)
        
        if i % 100 == 0:
            cost_list.append(cost)
            index_list.append(i)
            print ("Cost after iteration %i: %f" %(i, cost))
    plt.plot(index_list,cost_list)
    plt.xticks(index_list,rotation='vertical')
    plt.xlabel("Number of Iterarion")
    plt.ylabel("Cost")
    plt.show()
    
    # predict
    y_prediction_test = predict_NN(parameters,x_test)
    y_prediction_train = predict_NN(parameters,x_train)

    # Print train/test Errors
    print("train accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_train - y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_test - y_test)) * 100))
    return parameters

parameters = two_layer_neural_network(x_train, y_train,x_test,y_test, num_iterations=2500)
```

num_iterations parametresi öğrenme sürecinin uzunluğunu belirtir.

Sonrasında görselleştirip azaliz yapabilmek için cost_list array'inde cost'lar ve index_list array'inde indeksler depolanmıştır.

parameters = initialize_parameters_and_layer_sizes_NN(x_train, y_train) satırında yapılan, parametreleri initialize ederken w1, w2, b1, b2'nin boyutlarını belirtmek gerekir. Bu yüzden parametre olaran x_train ve y_train veriyoruz çünkü boyutları buradan çekebileceğiz.

Böylece artık elimizde random bir şekilde initialize edilmiş 0.01 civarında bulunan weight'ler ve bias'lar mevcut ve bunlar random yani birbirlerinden farklılar.

## Implementing with keras library

Artificial Neural network algoritmasını keras kütüphanesi kullanarak implement edelim.

```python
# Evaluating the ANN
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential # initialize neural network library
from keras.layers import Dense # build our layers library
def build_classifier():
    classifier = Sequential() # initialize neural network
    classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu', input_dim = x_train.shape[1]))
    classifier.add(Dense(units = 4, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
classifier = KerasClassifier(build_fn = build_classifier, epochs = 100)
accuracies = cross_val_score(estimator = classifier, X = x_train, y = y_train, cv = 3)
mean = accuracies.mean()
variance = accuracies.std()
print("Accuracy mean: "+ str(mean))
print("Accuracy variance: "+ str(variance))
```

build_classifier() fonksiyonu neural network'ü oluşturan yapıdır.

classifier = Sequential() # initialize neural network 

Bu satırda, Sequential() metodunu çağırarak neural network yapısı oluşturuluyor. Bu yapıya layer'lar eklenmelidir.


classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu', input_dim = x_train.shape[1]))

Yukarıdaki satırı inceleyelim. Dense yapısı layer'ları inşa etmemize yarar. units parametresi ile layer'daki output'ların boyutu belirtilir. Bu satırda hidden layer'da 8 tane node kullanılacağı belirtilmiştir. kernel_initializer, weight'leri initialize etmemize yarar ve burada uniform şekilde dağılım sağlanmıştır. activation = 'relu' parametresi ile aktivasyon fonksiyonunun relu olması sağlanmıştır.

Relu'nun ne olduğunu anlamak için sigmoid ile karşılaştıralım.

Sigmoid input'u 0 ile 1 arasında sınırlandırır. 

Relu, eğer input 0'dan küçük ise sonuç 0'dır fakat input 0'dan büyük ise output=input'tur. Mesela input=-5 ise output=0'dır. Mesela input=5 ise output=5'dır.

input_dim = x_train.shape[1] parametresi input size'ıdır. Burada 4096'dır.

classifier.add(Dense(units = 4, kernel_initializer = 'uniform', activation = 'relu'))

Bu satırda 2.hidden layer tanımlanmıştır.

classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

Bu satırda 3.hidden layer tanımlanmıştır. Aktivasyon fonksiyonunun sigmoid olması, son output layer'ının eklendiği anlamına gelir.

classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

Bu satırda classifier'ı compile etmek için gerekli olan optimizasyon parametreleri tanımlanmıştır. loss = 'binary_crossentropy' parametresi ile loss değeri hesaplanmıştır.

Bu işlem sonrasında backward propagation kullanılmalıdır. Learning rate'i sabit tutarak gradient descent yapıyorduk. Yine gradient descent yapacağız ama kullanacağımız optimizasyon "adam" olucak. "adam" ın anlamı adaptive momentum'dur. 

Gradient descent'te learning rate'i sabit tutuyorduk. "adam" da ise learning rate sabit olmuyor, belli bir momentuma adapte olarak learning rate'i güncelleyerek gidiyor. Yani learning rate'de değişiyor ve daha hızlı öğrenmeye neden oluyor. "adam" ile memory'in daha efektif kullanımı sağlanır.

metrics = ['accuracy'] parametresi değerlendirme metriğini ifade eder. Bir model oluşturuyoruz ve bu modeli değerlendirirken accuracy kullanacağımızı belirtiyoruz.

Ve son olarak oluşturduğumuz modeli yani classifier'ı return ediyoruz.

classifier = KerasClassifier(build_fn = build_classifier, epochs = 100)

Bu satırda oluşturduğumuz classifier'ı çağırılıp kullanılmıştır.Bunun için KerasClassifier isimli metoddan yararlananılmıştır. build_fn = build_classifier parametresi ile buid edilen neural network çağrılmaktadır. epochs = 100 parametresi ile iterasyon sayısı belirtilmiştir.

accuracies = cross_val_score(estimator = classifier, X = x_train, y = y_train, cv = 3)

Şimdi de eğitim yapılmıştır. Eğitim içinde cross valudation kullanılmıştır. Cross validation ile birden fazla accuracy elde ediliyor ve bu accuracy değerlerinin ortalaması alınarak tek bir accuracy değeri elde ediliyor ve ortalama işlemi ile daha efektif bir sonuç elde etmiş oluyoruz. cv = 3 parametresinden 3 tane accuracy değeri elde edileceğini anlıyoruz.

mean = accuracies.mean()

Bu satıra elde edilen 3 tane accuracy değerlerinin ortalaması alınmıştır.

variance = accuracies.std()

Bu satırda accuracy değerlerinin varyansı alınmıştır.
